```python
# importing networkx
import networkx as nx

# importing matplotlib.pyplot
import matplotlib.pyplot as plt

g = nx.Graph()

g.add_edge(1,21)
g.add_edge(1,22)
g.add_edge(1,23)
g.add_edge(1,24)
g.add_edge(1,25)
g.add_edge(21,3)
g.add_edge(22,3)
g.add_edge(23,3)
g.add_edge(24,3)
g.add_edge(25,3)
g.add_edge(3,41)
g.add_edge(3,42)
g.add_edge(3,43)
g.add_edge(3,44)
g.add_edge(3,45)

# drawing the plot
nx.draw(g)
plt.savefig("task2-1.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing the plot with numbering
nx.draw(g, with_labels = True)
plt.savefig("task2-2.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in circular layout
nx.draw_circular(g, with_labels = True)
plt.savefig("task2-3.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in planar layout
nx.draw_planar(g, with_labels = True)
plt.savefig("task2-4.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in random layout
nx.draw_random(g, with_labels = True)
plt.savefig("task2-5.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in spectral layout
nx.draw_spectral(g, with_labels = True)
plt.savefig("task2-6.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in spring layout
nx.draw_spring(g, with_labels = True)
plt.savefig("task2-7.png")
plt.show()
# clearing the current plot
plt.clf()

# drawing in shell layout
nx.draw_shell(g, with_labels = True)
plt.savefig("task2-8.png")
plt.show()
```
