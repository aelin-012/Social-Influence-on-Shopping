
<span style="color: WHITE; font-size: 30px;">BETWENNESS CENTRALITY

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph and add edges
g = nx.Graph()
edges = [(1, 21), (1, 22), (1, 23), (1, 24), (1, 25), (21, 3), (22, 3), (23, 3), (24, 3), (25, 3), (3, 41), (3, 42), (3, 43), (3, 44), (3, 45)]
g.add_edges_from(edges)

# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(g)

# Print the betweenness centrality for each node
for node, centrality in betweenness_centrality.items():
    print(f"Node {node}: Betweenness Centrality = {centrality}")

# Visualization (optional)
pos = nx.spring_layout(g)  # Layout for visualization
nx.draw(g, pos, with_labels=True, font_weight='bold', node_color='skyblue', node_size=800, font_size=8)
plt.show()
```
